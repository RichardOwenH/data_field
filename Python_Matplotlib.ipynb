### Example 2

#### Create Data

```python
''' list x : nilai dari 1 sampe 100
    list y1 : nilai pangkat 2 dari x
    list y2 : nilai pangkat 3 dari x
'''

def fungsi_pangkat(nilai_awal, nilai_akhir):
  x = np.arange(nilai_awal, nilai_akhir+1)
  y1 = np.power(x,2)
  y2 = np.power(x,3)
  return x,y1,y2
```

```python
x,y1,y2 = fungsi_pangkat(1,10)
```

#### Create Chart + Show

```python
# Ploting Our Data
plot1 = plt.plot(x, y1)
plot2 = plt.plot(x, y2)

# Show Chart
plt.show()
```

#### Add Properties

```python
# Ploting Our Data
plot1 = plt.plot(x, y1)
plot2 = plt.plot(x, y2)

# Add Properties setp(name_chart, color={warna line}, linestyle={tipe linenya}, linewidth={tebel linenya})
plt.setp(plot1, color='r', linestyle='--', linewidth=1.5)
plt.setp(plot2, color='skyblue', linestyle='dashed', linewidth=0.8)
# '-', '--','dashed',':','dashdot',...

# Show Chart
plt.show()
```

#### Add Label

```python
# Ploting Our Data
plot1 = plt.plot(x, y1)
plot2 = plt.plot(x, y2)

# Add Properties setp(name_chart, color={warna line}, linestyle={tipe linenya}, linewidth={tebel linenya})
plt.setp(plot1, color='r', linestyle='--', linewidth=1.5)
plt.setp(plot2, color='skyblue', linestyle='dashed', linewidth=0.8)
# '-', '--','dashed',':','dashdot',...

# Set Label
plt.xlabel('Nilai')
plt.ylabel('Nilai Pangkat')
plt.title('Plot Nilai Pangkat')

# Show Chart
plt.show()
```

#### Set Axis

```python
# Ploting Our Data
plot1 = plt.plot(x, y1)
plot2 = plt.plot(x, y2)

# Add Properties setp(name_chart, color={warna line}, linestyle={tipe linenya}, linewidth={tebel linenya})
plt.setp(plot1, color='r', linestyle='--', linewidth=1.5)
plt.setp(plot2, color='skyblue', linestyle='dashed', linewidth=0.8)
# '-', '--','dashed',':','dashdot',...

# Set Label
plt.xlabel('Nilai')
plt.ylabel('Nilai Pangkat')
plt.title('Plot Nilai Pangkat')

# Set Axis axis([x_min, x_max, y_min, y_max])
plt.axis([1,10, 1, 1200])
# kalo di math, standarnya mesti dari 0

# Show Chart
plt.show()
```

#### Add Legend

```python
# Ploting Our Data
plot1 = plt.plot(x, y1, label='Nilai Pangkat 2')
plot2 = plt.plot(x, y2, label='Nilai Pangkat 3')

# Add Properties setp(name_chart, color={warna line}, linestyle={tipe linenya}, linewidth={tebel linenya})
plt.setp(plot1, color='r', linestyle='--', linewidth=1.5)
plt.setp(plot2, color='skyblue', linestyle='dashed', linewidth=0.8)
# '-', '--','dashed',':','dashdot',...

# Set Label
plt.xlabel('Nilai')
plt.ylabel('Nilai Pangkat')
plt.title('Plot Nilai Pangkat')

# Set Axis axis([x_min, x_max, y_min, y_max])
plt.axis([1,10, 1, 1200])
# kalo di math, standarnya mesti dari 0

# Add Legend
#plt.legend(loc='upper left')
plt.legend(loc='upper left', bbox_to_anchor=(1,1.05))

# Show Chart
plt.show()
```

#### Add Text

```python
# Ploting Our Data
plot1 = plt.plot(x, y1, label='Nilai Pangkat 2')
plot2 = plt.plot(x, y2, label='Nilai Pangkat 3')

# Add Properties setp(name_chart, color={warna line}, linestyle={tipe linenya}, linewidth={tebel linenya})
plt.setp(plot1, color='r', linestyle='--', linewidth=1.5)
plt.setp(plot2, color='skyblue', linestyle='dashed', linewidth=0.8)
# '-', '--','dashed',':','dashdot',...

# Set Label
plt.xlabel('Nilai')
plt.ylabel('Nilai Pangkat')
plt.title('Plot Nilai Pangkat')

# Set Axis axis([x_min, x_max, y_min, y_max])
plt.axis([1,10, 1, 1200])
# kalo di math, standarnya mesti dari 0

# Add Legend
#plt.legend(loc='upper left')
plt.legend(loc='upper left', bbox_to_anchor=(1,1.05))

# Add Text text(posisi x, posisi y, 'textnya apa')
plt.text(8,90,r'${x}^2$')
plt.text(8,590,r'${x}^3$')

# Show Chart
plt.show()
```

## Type of Chart

### Histogram

untuk menampilkan data numerik yg kontinu

```python
# Create Data
## kita mau generate data dist. normal dgn total data 1000
center, sigma = 100, 15
x = center + sigma*np.random.randn(1000)
x.shape

# Create Chart hist(data, bins={total group data}, facecolor={warna chart}, alpha={berapa kontras warna})
plot1 = plt.hist(x, bins=25, facecolor='r', alpha=0.5)

# Set Label
plt.xlabel('Nilai x')
plt.ylabel('Total Data')
plt.title('Histogram of Data')

# Add Grid
plt.grid()

# Show Chart
plt.show()
```

### Bar Chart

buat data diskrit (kategori & total)

```python
# Create Data
x = ['FOOD','RIDE','CAR','CLEAN','SHOP']
y1 = [1000, 1500, 800, 400, 600]

# Create Chart
plot1 = plt.bar(x, y1, color='blue', alpha=0.5)

# Set Label
plt.xlabel('Category')
plt.ylabel('Total Orders')
plt.title('Total Orders by Category')

# Show Chart
plt.show()
```

### Horizontal Bar Chart

```python
# Create Data
x = ['FOOD','RIDE','CAR','CLEAN','SHOP']
y1 = [1000, 1500, 800, 400, 600]

# Create Chart
plot1 = plt.barh(x, y1, color='blue', alpha=0.5)

# Set Label
plt.xlabel('Category')
plt.ylabel('Total Orders')
plt.title('Total Orders by Category')

# Show Chart
plt.show()
```

### Stacked Bar

```python
# Create Data
x = ['FOOD','RIDE','CAR','CLEAN','SHOP']
y1 = [1000, 1500, 800, 400, 600]
y2 = [800, 1300, 850, 600, 700]

# Create Chart
plot1 = plt.bar(x, y1, color='blue', alpha=0.5, label='apps')
plot2 = plt.bar(x, y2, color='red', alpha=0.5, bottom=y1, label='web')

# Set Label
plt.xlabel('Category')
plt.ylabel('Total Orders')
plt.title('Total Orders by Category')

# Add Legend
plt.legend()

# Show Chart
plt.show()
```

### Grouped Bar Chart

```python
# Create Data
x = ['FOOD','RIDE','CAR','CLEAN','SHOP']
y1 = [1000, 1500, 800, 400, 600]
y2 = [800, 1300, 850, 600, 700]

x_arange = np.arange(len(x))
width = 0.3 #lebar batang chartnya (0.0 - 1.0)

# Create Chart
plot1 = plt.bar(x_arange-width/2, y1, width, color='blue', alpha=0.5, label='apps')
plot2 = plt.bar(x_arange+width/2, y2, width, color='red', alpha=0.5, label='web')

# Set Ticks
plt.xticks(x_arange, x)

# Set Label
plt.xlabel('Category')
plt.ylabel('Total Orders')
plt.title('Total Orders by Category')

# Add Legend
plt.legend()

# Show Chart
plt.show()
```

### Pie Chart
1 kategori, 1 value 

```python
# Create Data
x           = ['FOOD','RIDE','CAR','CLEAN','SHOP']
y1          = [1000,    1500,  800,    400,  600]
explode_var = [0.,       0.1,   0.,     0.,   0.]

# Create Chart
plot1 = plt.pie(y1, labels=x, autopct='%1.1f%%', explode=explode_var)

# Set Label
plt.title('Total Orders by Category')

# Add Legend
plt.legend(loc='upper right', bbox_to_anchor=(1.35,1))

# Show Chart
plt.show()
```

### Scatter Plot

```python
# Create Data
x = [1,4,5,7,9]
y = [2,5,7,8,10]

x1 = [1,3,5,6,12]
y1 = [2,7,6,9,13]

# Create Chart
plot1 = plt.scatter(x, y, label='Data 1', color='g')
plot2 = plt.scatter(x1, y1, label='Data 2', color='r')

# Set Label
plt.xlabel('variable 1')
plt.ylabel('variable 2')
plt.title('Relation between var1 and var2')

plt.legend()

# Show Chart
plt.show()

# variabel 1 dan variabel 2 berkorelasi positif 
```

### Boxplot

```python
# Create Data
x = np.random.normal(loc=80, scale=10, size=250)
y = np.random.normal(loc=70, scale=20, size=250)

data = [x,y]

# Create Chart
plot1 = plt.boxplot(data, labels=['data x', 'data y'], showfliers=True, vert=False)

# Show Chart
plt.show()
```
